```python
data = pd.read_csv("C:/Users/kanwe/Downloads/Catapult.csv")
data.head()
```

```python
# Use the column names as they appear in the dataset (with spaces)
lat_lon_data = data[[' Latitude', ' Longitude']]
```

```python
print(lat_lon_data)
```

```python
import numpy as np  # Import numpy
import seaborn as sns
import matplotlib.pyplot as plt

# Create a 2D histogram (heatmap) of player positions
heatmap, xedges, yedges = np.histogram2d(lat_lon_data[' Longitude'], lat_lon_data[' Latitude'], bins=100)
```

```python
# Plot the heatmap using seaborn and matplotlib
# Plot the heatmap using seaborn and matplotlib
plt.figure(figsize=(10, 6))
sns.heatmap(heatmap.T, cmap='coolwarm', cbar=True)

# Set axis labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Player Position Heatmap')

# Display the heatmap
plt.show()
```
